```python
check = np.zeros((9,9))
# 9 x 9 의 행렬
```

```python
check[::2,1::2] = 1
# 행기준 짝수인 경우 흰색이 발생 
check[1::2, ::2] =1
# 열기준 짝수인 경우 흰색이 발생
```

```python
plt.imshow(check,cmap='gray',interpolation='nearest')
```
